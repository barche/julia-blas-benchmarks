````julia
using Plots
````

````julia
function analyze(ob,obpar,mklpar,mklser,steps,i)
    plot(n,ob[:,i],label="OpenBLAS serial",color=:blue)
    plot!(n,obpar[:,i],label="OpenBLAS threads",style=:dot,color=:blue)
    plot!(n,mklpar[:,i],label="MKL threads",style=:dash,color=:red)
    plot!(n,mklser[:,i],label="MKL serial",color=:red)
    title!(steps[i])
end
````

````julia
pyplot()
````
